# Korelacja i regresja liniowa
## Korelajca
Korelacja jest to związek pomiedzy dwoma zmiennymi. Do sprawdzania korelacji liniowych używamy współczynnika pearsona. Przyjmuje on wartość od -1 do 1. Czym wartość bezwzgledna jest bliższa 1 tym bardziej dane są skorelowane. Wartość 0 oznacza brak korelacji. Wartość ujemna oznacza korelacje ujemną, a dodatnia dodatnią.
Wzór na współczynnik pearsona:
$$r = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n}(x_i - \bar{x})^2}\sqrt{\sum_{i=1}^{n}(y_i - \bar{y})^2}}$$
**Oto przykłady:**

```r
n <- 100
t <- seq(0,4,0.2)

data1 <- data.frame(n, x <- runif(n),1.5*x+runif(n)/2)
data2 <- data.frame(n, sin(t), cos(t))
data3 <- data.frame(n, runif(n), runif(n))
data4 <- data.frame(n, -x, 1.5*x+runif(n)/2)
par(mfrow=c(2,2))
print(cor(data1[2:3])[2])
plot(data1[2:3], main=cor(data1[2:3])[2])
plot(data2[2:3], main=cor(data2[2:3])[2])
plot(data3[2:3], main=cor(data3[2:3])[2])
plot(data4[2:3], main=cor(data4[2:3])[2])
```

```r
correlation_pearson <- function(x, y){
    n <- length(x)
    x_bar <- mean(x)
    y_bar <- mean(y)
    s_x <- sd(x)
    s_y <- sd(y)
    sum((x-x_bar)*(y-y_bar))/sqrt(sum((x-x_bar)^2)*sum((y-y_bar)^2))
}
x = data1[,2]
y = data1[,3]
correlation_pearson(x, y)
```

## Kowariancja
Kowariancja jest to miara zależności liniowej między dwiema zmiennymi losowymi z tą różnicą że nie jest ona normalizowana oraz nie informuje o sile oraz kierunku zależności. Wzór na kowariancję:
$$cov(X,Y) = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{n}$$
**Oto przykłady:**

```r
par(mfrow=c(2,2))
plot(data1[2:3], main=cov(data1[2:3])[2])
plot(data2[2:3], main=cov(data2[2:3])[2])
plot(data3[2:3], main=cov(data3[2:3])[2])
plot(data4[2:3], main=cov(data4[2:3])[2])
```

```r
covariancy <- function(x, y){
    n <- length(x)
    x_bar <- mean(x)
    y_bar <- mean(y)
    sum((x-x_bar)*(y-y_bar))/n
}
x = data1[,2]
y = data1[,3]
covariancy(x, y)
```

## Regresja Liniowa
Regresja liniowa to metoda wyznaczania zależności między zmiennymi. W przypadku regresji liniowej zależność ta jest liniowa oraz wyznacza ona wzór funkcji liniowej. Obliczana jest za pomocą algorytmu najmniejszych kwadratów. Polega ona na tym że bierzemy średnią wartość z każdych wartości. Średnie wartości tworzą punkt przez który bedzie musiała przebiegać nasza funkcja.

```r
plot(data1[2:3])
points(mean(data1[,2]), mean(data1[,3]), col="red", pch=19)
```

Następnie szukamy współczynników a oraz b funkcji liniowej y = ax + b które maja najmniejszą różnice kwadratów pomiedzy wszystkimi punktami. Wzór na regeresje liniowa
$$y = ax + b$$
$$a = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}$$
$$b = \bar{y} - a\bar{x}$$

```r
plot(data1[2:3])
points(mean(data1[,2]), mean(data1[,3]), col="red", pch=19)
abline(lm(data1[,3]~data1[,2]), col="blue")
print(lm(data1[,3]~data1[,2])$coefficients)
```

```r
line_regresion <- function(x,y){
    n <- length(x)
    x_bar <- mean(x)
    y_bar <- mean(y)
    b <- sum((x-x_bar)*(y-y_bar))/sum((x-x_bar)^2)
    a <- y_bar-b*x_bar
    c(a,b)
}
line_regresion(x,y)
```
