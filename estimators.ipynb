# Estymatory
Estymatory to są funkcje, które przyjmują próbkę danych i zwracają wartość, która jest przybliżeniem parametru populacji. W przypadku próbki z rozkładu normalnego, estymator średniej populacji jest średnią próbkową. W przypadku próbki z rozkładu normalnego, estymator wariancji populacji jest wariancją próbkową.
## Estymator nieobciążony
Estymator nieobciążony to taki którego wartość estymatora średnio jest równa wartości parametru populacji.
## Estymator obciążony 
Estymator obciążony to taki którego wartość średnio oscyluje wokół wartości parametru populacji.
## Badanie obciążenia estymatorów
### Losowanie populacji

```r
population_size <- 1000
population <- runif(population_size, 0, 20)
```

### Losowanie próby

```r
sample_size <- 100
sample_count <- 200
samples <- replicate(sample_count, sample(population, sample_size))
samples
```

### Statystyki opisowe dla populacji

```r
variance_population <- function(data){
    var(data)/(length(data)/(length(data)-1))
}

descriptive_statistics_population <- c(
    mean = mean(population),
    variance = variance_population(population),
    sd = sqrt(variance_population(population))
)
descriptive_statistics_population
```

### Estymatory

```r
estimators <- list(
    mean_unbiased = apply(samples, 2, mean),
    variance_unbiased = apply(samples, 2, variance_population),
    variance_biased = apply(samples, 2, var)
)
```

### Porównanie Estymatorów 

```r
rows  <- c("Estimator Mean", "Biased")
columns  <- c("Estimator mean", "Estimator variance", "Estimator biased variance")

average_estimators = sapply(estimators, mean)

biaseds = c(
    descriptive_statistics_population["mean"]- average_estimators["mean_unbiased"],
    descriptive_statistics_population["variance"]- average_estimators["variance_unbiased"],
    descriptive_statistics_population["variance"]- average_estimators["variance_biased"]
)
matrix(c(rbind(average_estimators, biaseds)), 2, 3, dimnames=list(rows, columns))
```

### Wartości estymatorów

```r
# get value by  name from list
hist(estimators$mean_unbiased)
hist(estimators$variance_unbiased)
hist(estimators$variance_biased)
```

## Estymatory przedziałowe

```r
alfa <- 0.05
mean <- descriptive_statistics_population["mean"]
sd <- descriptive_statistics_population["sd"]
quantile <- qnorm(1- alfa/2, 0, 1)
left_interval <- average_estimators + (quantile*sqrt(sd^2/sample_size))
right_interval <- average_estimators - (quantile*sqrt(sd^2/sample_size))
matrix(
    c(rbind(left_interval, right_interval, left_interval - right_interval, average_estimators, c(descriptive_statistics_population["mean"], descriptive_statistics_population["variance"], descriptive_statistics_population["variance"]))),
    5, 
    3, 
    dimnames=list(c("Left interval", "Right interval", "Interval long", "Average estimators", "Real values"), columns))
```

### Estymator przedziałowy na wykresie

```r
draw_ci_for_statistic = function(statistic, left_lim, right_lim, real, to=20){
    lim <- c(left_lim, right_lim)
    curve(dnorm(x, real, 1), ylab = "Density", from=0, to=to, xlab = "x")
    abline(h = dnorm(real, real, 1) * c(0, 1), col = "blue", lty = c(1, 2))
    x <- seq(lim[1], lim[2], length = 100)
    y <- dnorm(x, real, 1)
    polygon(c(lim[1], x, lim[2]), c(0, y, 0), col = "lightblue")
    abline(v = statistic, col = "red")
}
draw_ci_for_statistic(average_estimators["mean_unbiased"], left_interval["mean_unbiased"], right_interval["mean_unbiased"], descriptive_statistics_population["mean"])
draw_ci_for_statistic(average_estimators["variance_unbiased"], left_interval["variance_unbiased"], right_interval["variance_unbiased"], descriptive_statistics_population["variance"], 40)
draw_ci_for_statistic(average_estimators["variance_biased"], left_interval["variance_biased"], right_interval["variance_biased"], descriptive_statistics_population["variance"], 40)
```
